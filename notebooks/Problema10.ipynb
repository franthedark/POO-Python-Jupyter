```python
#clase Empleado
class Empleado:
    def __init__(self, nombre: str, salario_base: float):
        self.nombre = nombre            #nombre del empleado
        self.salario_base = salario_base  #salario base mensual

    def calcular_pago(self) -> float:
        """
        Retorna el salario base mensual del empleado.
        """
        return self.salario_base

    def __repr__(self):
        return f"<Empleado: {self.nombre}, Salario base: {self.salario_base:.2f}>"

#clase Nomina
class Nomina:
    def __init__(self):
        self.empleados = []  #lista de objetos Empleado

    def agregar_empleado(self, empleado: Empleado):
        """
        Agrega un objeto Empleado a la nómina.
        """
        self.empleados.append(empleado)

    def calcular_total(self) -> float:
        """
        Retorna la suma de los pagos de todos los empleados.
        """
        return sum(e.calcular_pago() for e in self.empleados)

    def __repr__(self):
        return f"<Nómina: {len(self.empleados)} empleados, Total: {self.calcular_total():.2f}>"

#clase EmpleadoConBono (hereda de Empleado)
class EmpleadoConBono(Empleado):
    def __init__(self, nombre: str, salario_base: float, bono: float):
        super().__init__(nombre, salario_base)
        self.bono = bono  #ingreso adicional mensual

    def calcular_pago(self) -> float:
        """
        Retorna el salario base más el bono.
        """
        return self.salario_base + self.bono

    def __repr__(self):
        return (f"<EmpleadoConBono: {self.nombre}, Salario base: {self.salario_base:.2f}, "
                f"Bono: {self.bono:.2f}>")
```

```python
#celda de demostracion
#crear empleados sin bono
e1 = Empleado("Ana", 1000.00)
e2 = Empleado("Luis", 1200.00)

#crear empleados con bono
e3 = EmpleadoConBono("María", 1100.00, 200.00)
e4 = EmpleadoConBono("Pedro", 900.00, 150.00)

#crear nomina y agregar empleados
nomina = Nomina()
nomina.agregar_empleado(e1)
nomina.agregar_empleado(e2)
nomina.agregar_empleado(e3)
nomina.agregar_empleado(e4)

#mostrar pagos individuales
print(e1, "→ Pago:", e1.calcular_pago())
print(e2, "→ Pago:", e2.calcular_pago())
print(e3, "→ Pago:", e3.calcular_pago())
print(e4, "→ Pago:", e4.calcular_pago())

#calcular y mostrar total de la nomina
print(f"\nTotal nómina mensual: {nomina.calcular_total():.2f}")
print(nomina)
```
